```python
##Responder: ¿Cuál archivo tiene el mayor número de palabras? y su contenido
```

```python
filename = ('A un general (Julio Corta╠üzar).txt')
try:
    with open('archivos/poemas/' + filename) as f_obj:
        contents = f_obj.read()
        print(contents)
except FileNotFoundError:
    msg = "El archivo" + filename  + "No se encontró"
    print(msg)
    
else:
    #contar palabras
    words =  contents.split() #convierte a una lisya
    num_words = len(words) #lee la lista
    print("El poema " + filename + " contiene " + str(num_words) + " palabras")
```
