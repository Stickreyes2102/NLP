```python
#Punto 1:
#1.1 Leer el archivo UNA_SOLITARIA_VOZ_HUMANA.txt con with open
```

```python
##Importar Libreria
import bs4 as bs
```

```python
# Abrir archivo en .txt en Python
with open("archivos/UNA_SOLITARIA_VOZ_HUMANA.txt","r",encoding='utf-8') as archivo:
    data = archivo.read()
    print(data)
    
```

```python
# 1.2 Convertir a minúsculas
```

```python
data = data.lower() # Minúscula
data
```

```python
#1.3 Conservar sólo palabras
```

```python
import re
data = re.sub(r"[\W\d_]+", " ", data)
data
```

```python
# 1.4 Tokenizar
```

```python
data = data.split() # Tokenización
data
```

```python
#Punto 2:
#1.1 Quitar palabras vacias
```

```python
import nltk
#nltk.download('stopwords')

from nltk.corpus import stopwords
stopwords_sp = stopwords.words('spanish')

data = [palabra for palabra in data if palabra not in stopwords_sp]
data
```

```python
# 2.1¿Cuáles son las 10 palabras no vacias más usadas?
```

```python
contador = {}
for palabra in data:
    if palabra not in contador:
        contador[palabra] = 1
    else:
        contador[palabra] += 1
```

```python
print('Las diez palabras más comunes son:')
for palabra in sorted( contador, key = contador.get, reverse=True )[:10]:
    print('\t',palabra,':',contador[palabra],'ocurrencias') 
```

```python
print('Total de palabras distintas:', len(data))        
```

```python
#Punto 3:
#Stemming del documento
```

```python
from nltk.stem.snowball import SnowballStemmer
#englishStemmer=SnowballStemmer("english")
spanishStemmer=SnowballStemmer("spanish")

Stemming = []
for palabra in data: 
    raiz = spanishStemmer.stem(palabra)
    Stemming.append(raiz)
```

```python
#¿Cuáles son las 10 raíces más usadas?
```

```python
contador2 = {}
for palabra in Stemming:
    if palabra not in contador2:
        contador2[palabra] = 1
    else:
        contador2[palabra] += 1
```

```python
print('Las diez raíces más comunes son:')
for palabra in sorted( contador2, key = contador2.get, reverse=True )[:10]:
    print('\t',palabra,':',contador2[palabra],'ocurrencias') 
```
