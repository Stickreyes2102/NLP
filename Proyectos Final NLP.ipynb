####  *Se crea una función que toma como argumento un link, 
#### *Parseamos el html
#### *Guarda el titulo de la canción en el objeto title
#### *Arma un diccionario en la variable songs donde tiene dos elementos (titulo y letra)


```python
def get_song(song_url):

    links_ = requests.get(song_url)  
    links_ = links_.content  
    soup_ = BeautifulSoup(links_)
    
    title = soup_.find('h3').get_text()#obtenemos los títulos, que son un elemento único en el tag h3
    
    lyrics = []
    for i in soup_.find('div', {'id':'letra'}).find_all('p'):
        if i.has_attr('class'): #filtramos los elementos p que no son estrofas
            break
            
        elif i.a: #filtramos los elementos p que no son estrofas
            pass
        else:
            lyrics.append((i.get_text(strip=True, separator=' ')))
            
    lyrics = ' '.join([str(elem) for elem in lyrics])
    song = {'title':title, 'lyrics':lyrics}
    return song
    
                
```

```python
songs = []
for l in links:
    print(l)
    songs.append(get_song(l))
    
```

```python
vocabulario = df.preprocesado.str.cat(sep=" ")##concatenar todos los textos
vocabulario = set(vocabulario.split())##quitar elementos repetidos en la lista con set
vocabulario = [(word,i+1) for i,word in enumerate(vocabulario)]##enumerar cada palabra de la lista comenzando desde 1
print(vocabulario)
```

```python
##Bolsa de palabras
def bow(frase):
    frase_bow = [0]*len(vocabulario)
    for word,i in vocabulario:
        if word in frase.split():
            frase_bow[i-1] = 1
    print(frase_bow)
bow("baby enteré nota ve ahí llega sabe llevaré dim.")
```

```python
#Recolectar información: Tweets de la cuidad de Bogotá
tweets = tweepy.Cursor(api.search,
                       q=['BadBunny -RT'],
                       lang='es',
                       geocode=f'4.60971,-74.08175,45mi',
                       tweet_mode='extended',
                       include_rts=False).items(5000)
```

```python
t = []

for tweet in tweets:
    temp = []
    temp.append(tweet.full_text)
    temp.append(tweet.created_at)
    temp.append("Bogota")
    temp.append(tweet.user.location)
    temp.append(tweet.id)
    temp.append(f"https://twitter.com/i/web/status/{tweet.id}")
    
    t.append(temp)
```

```python
data = pd.DataFrame(t)
data.columns = ['tweet', 'date', 'city', 'user_location', 'id', 'link']
data.head()
```

```python
data = data[['tweet', 'date', 'city']]
data.head()
```

```python
# Realizamos el Preprocesamiento de la data: Se depurar emojis, hashtags, hipervinculos
pattern1 = '?P<pic>pic.twitter.com/[^\s]+'
pattern2 = '?P<url>https?://[^\s]+'

def text_clean(row):
    text = row['tweet']
    
    links = [tuple(j for j in i if j)[-1] for i in re.findall(f"({pattern1})|({pattern2})",text)]
    for link in links:
        text = text.replace(link,"")
             
    hashtags = [interaction for interaction in text.split() if interaction.startswith("#")]
    for hashtag in hashtags:
        text = text.replace(hashtag,"")
        
    mentions = [interaction for interaction in text.split() if interaction.startswith("@")]
    for mention in mentions:
        text = text.replace(mention,"")
        
    return text, links, hashtags, mentions
```

```python
data[['clean', 'links', 'hashtags', 'mentions']] = data.apply(text_clean, axis=1, result_type='expand')
data.head()
```

```python
import regex
import emoji

def get_emojis(text):
    emoji_list = []
    data = regex.findall(r'\X', text)
    for word in data:
        if any(char in emoji.UNICODE_EMOJI for char in word):
            emoji_list.append(word)

    return emoji_list

data['emojis'] = data['clean'].apply(lambda text: get_emojis(text))
data.head()
```
