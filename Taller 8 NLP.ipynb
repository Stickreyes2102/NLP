```python
import re
import pandas as pd 
from pprint import pprint

from nltk.corpus import stopwords
stopwords = stopwords.words('spanish')

import pyLDAvis.gensim
from gensim.models import LdaModel
from gensim.corpora import Dictionary

import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
# Cargar datos
path = 'archivos/reviews_vidjew_es.csv'
data = pd.read_csv(path)
data.head()
```

```python
###  `[12 pts]` Punto 1: Hacer pre-procesamiento del texto
import re
from nltk.corpus import stopwords
stopwords = stopwords.words('spanish')
```

```python
def pre_procesado(texto):
    texto = texto.lower()

    texto = re.sub(r"[\W\d_]+", " ", texto)
    texto = [palabra for palabra in texto.split() if palabra not in stopwords]
    return texto
 
data['pp'] = data['review_body'].apply(lambda texto: pre_procesado(texto))
data.head()
```

```python
###  `[13 pts]` Punto 2: Modelo de LDA
```

```python
# Crear una representaci贸n de los documentos en forma de diccionario

#Cargamos librerias para formar el diccionario, procesar el modelo lda y quitar los warning cuando se usan las 
#import pyLDAvis.gensim
#from gensim.models import LdaModel
#from gensim.corpora import Dictionary
#from pprint import pprint
#import warnings
#warnings.filterwarnings("ignore", category=DeprecationWarning)


dictionary = Dictionary(data['pp'].values)
 
```

```python
# Filtrar palabras muy frecuentes y poco frecuentes
dictionary.filter_extremes(no_below=5, no_above=0.5)
corpus = [dictionary.doc2bow(text) for text in data['pp'].values]
```

```python

# Train the topic model 
model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=7, passes=20)
```

```python
#Resultados
model.print_topics(num_words=10)
```

```python
###  `[25 pts]` Punto 3: Visualizaci贸n de LDA
lda_display = pyLDAvis.gensim.prepare(model, corpus, dictionary, sort_topics=True)
pyLDAvis.display(lda_display)
```
